```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "E:/icu_project/models/ui_backend_data_v2.csv"
)

df.head()
```

```python
medications = [
    "Norepinephrine",
    "Epinephrine",
    "Propofol",
    "Fentanyl",
    "Midazolam",
    "Dobutamine",
    "Vasopressin"
]
```

```python
def generate_medication_events(prob):

    events = []

    if prob > 0.7:
        events.append("Norepinephrine")

    if prob > 0.6:
        events.append("Vasopressin")

    if prob > 0.5:
        events.append("Propofol")

    if prob > 0.4:
        events.append("Fentanyl")

    if prob < 0.3:
        events.append("Midazolam")

    return events
```

```python
df["medications"] = df["predicted_probability"].apply(
    generate_medication_events
)

df.head()
```

```python
timeline = [-4,-3,-2,-1,0]

def medication_timeline(prob):

    meds = generate_medication_events(prob)

    return {
        t: meds for t in timeline
    }
```

```python
df["medication_timeline"] = df["predicted_probability"].apply(
    medication_timeline
)
```

```python
df.to_csv(
    "E:/icu_project/models/ui_final_backend_dataset.csv",
    index=False
)
```
