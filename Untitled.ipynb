```python
import pandas as pd

selected_labs = [
    "WBC x 1000",
    "creatinine",
    "BUN",
    "sodium",
    "potassium",
    "magnesium",
    "Hct",
    "platelets x 1000",
    "Hgb",
    "HCO3",
    "pH"
]

labs = pd.read_csv(
    "data/eicu_demo/lab.csv",
    usecols=["patientunitstayid", "labresultoffset", "labname", "labresult"]
)

labs_filtered = labs[labs["labname"].isin(selected_labs)]

labs_filtered["labresult"] = pd.to_numeric(
    labs_filtered["labresult"],
    errors="coerce"
)

labs_filtered = labs_filtered.dropna(subset=["labresult"])

labs_wide = labs_filtered.pivot_table(
    index=["patientunitstayid", "labresultoffset"],
    columns="labname",
    values="labresult",
    aggfunc="mean"
).reset_index()

labs_wide.head()
```

```python
labs_wide = labs_wide.rename(columns={"labresultoffset": "offset"})
vitals = vitals.rename(columns={"observationoffset": "offset"})
```

```python
icu_merged = pd.merge(
    vitals,
    labs_wide,
    on=["patientunitstayid", "offset"],
    how="outer"
)

icu_merged = icu_merged.sort_values(
    ["patientunitstayid", "offset"]
)

print(icu_merged.shape)
icu_merged.head()
```

```python
icu_merged["hour"] = (icu_merged["offset"] // 60).astype(int)
```

```python
icu_hourly = icu_merged.groupby(
    ["patientunitstayid", "hour"]
).mean().reset_index()

print(icu_hourly.shape)
icu_hourly.head()
```

```python
# Drop raw minute offset
icu_no_offset = icu_merged.drop(columns=["offset"])

# Recompute hourly aggregation
icu_hourly = icu_no_offset.groupby(
    ["patientunitstayid", "hour"]
).mean().reset_index()

print(icu_hourly.shape)
icu_hourly.head()
```

```python
icu_hourly = icu_hourly.sort_values(
    ["patientunitstayid", "hour"]
)

icu_hourly = icu_hourly.groupby("patientunitstayid").ffill()

icu_hourly.head()
```

```python
missing_percent = icu_hourly.isna().mean() * 100
missing_percent.sort_values(ascending=False)
```

```python
drop_cols = [
    "temperature",
    "systemicsystolic",
    "systemicdiastolic",
    "systemicmean"
]

icu_hourly = icu_hourly.drop(columns=drop_cols)
```

```python
icu_hourly = (
    icu_merged
    .drop(columns=["offset"])
    .groupby(["patientunitstayid", "hour"], as_index=False)
    .mean()
)

print(icu_hourly.columns)
print(icu_hourly.shape)
```

```python
# Rebuild hourly dataset properly from icu_merged

icu_hourly = (
    icu_merged
    .assign(hour=(icu_merged["offset"] // 60).astype(int))
    .drop(columns=["offset"])
    .groupby(["patientunitstayid", "hour"], as_index=False)
    .mean()
)

print(icu_hourly.columns)
print(icu_hourly.shape)
```

```python
icu_hourly = (
    icu_hourly
    .sort_values(["patientunitstayid", "hour"])
    .groupby("patientunitstayid", group_keys=False)
    .apply(lambda x: x.ffill().bfill())
)

print(icu_hourly.columns)
```

```python
(icu_hourly.isna().mean() * 100).sort_values(ascending=False)
```

```python
drop_cols = [
    "temperature",
    "systemicsystolic",
    "systemicdiastolic",
    "systemicmean"
]

icu_hourly = icu_hourly.drop(columns=drop_cols)

(icu_hourly.isna().mean() * 100).sort_values(ascending=False)
```
