```python
df["deterioration_24h"] = (df["predicted_probability"] * 100).round(1)

df.head()
```

```python
def heart_status(p):

    if p < 0.3:
        return "Normal"
    elif p < 0.6:
        return "Stress"
    else:
        return "Critical"


def lung_status(p):

    if p < 0.35:
        return "Normal"
    elif p < 0.65:
        return "Improving"
    else:
        return "Compromised"


def kidney_status(p):

    if p < 0.4:
        return "Stable"
    elif p < 0.7:
        return "Monitor"
    else:
        return "Declining"


df["heart_status"] = df["predicted_probability"].apply(heart_status)
df["lung_status"] = df["predicted_probability"].apply(lung_status)
df["kidney_status"] = df["predicted_probability"].apply(kidney_status)
```

```python
def cardiovascular_status(row):

    if row["MAP"] >= 65 and row["HR"] < 100:
        return "Normal"
    elif row["MAP"] >= 60:
        return "Monitor"
    else:
        return "Shock Risk"


def respiratory_status(row):

    if row["SaO2"] >= 95 and row["RespRate"] < 22:
        return "Normal"
    elif row["SaO2"] >= 90:
        return "Compromised"
    else:
        return "Critical"


def renal_status(row):

    if row["Creatinine"] < 1.2:
        return "Stable"
    elif row["Creatinine"] < 2:
        return "Monitor"
    else:
        return "Failure Risk"


def neurologic_status(row):

    if row["GCS"] >= 14:
        return "Normal"
    elif row["GCS"] >= 10:
        return "Reduced"
    else:
        return "Severe"


def hepatic_status(row):

    if row["Bilirubin"] < 1.2:
        return "Normal"
    elif row["Bilirubin"] < 3:
        return "Impaired"
    else:
        return "Failure Risk"


def metabolic_status(row):

    if row["Lactate"] < 2:
        return "Stable"
    elif row["Lactate"] < 4:
        return "Stress"
    else:
        return "Critical"


def hematologic_status(row):

    if row["Platelets"] > 150:
        return "Normal"
    elif row["Platelets"] > 100:
        return "Monitor"
    else:
        return "Risk"
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("E:/icu_project/models/test_predictions.csv")

print(df.head())
print("Patients:", len(df))
```

```python
df["deterioration_24h"] = (df["predicted_probability"] * 100).round(1)
```

```python
def cardiovascular(p):
    if p < 0.3:
        return "Normal"
    elif p < 0.6:
        return "Stress"
    else:
        return "Critical"

def respiratory(p):
    if p < 0.35:
        return "Normal"
    elif p < 0.65:
        return "Compromised"
    else:
        return "Critical"

def renal(p):
    if p < 0.4:
        return "Stable"
    elif p < 0.7:
        return "Monitor"
    else:
        return "Failure Risk"

def neurologic(p):
    if p < 0.3:
        return "Normal"
    elif p < 0.6:
        return "Reduced"
    else:
        return "Severe"

def metabolic(p):
    if p < 0.35:
        return "Stable"
    elif p < 0.65:
        return "Stress"
    else:
        return "Critical"
```

```python
df["cardiovascular"] = df["predicted_probability"].apply(cardiovascular)
df["respiratory"] = df["predicted_probability"].apply(respiratory)
df["renal"] = df["predicted_probability"].apply(renal)
df["neurologic"] = df["predicted_probability"].apply(neurologic)
df["metabolic"] = df["predicted_probability"].apply(metabolic)
```

```python
def alert_level(p):

    if p < 0.25:
        return "Stable"
    elif p < 0.5:
        return "Watch"
    elif p < 0.75:
        return "Warning"
    else:
        return "Critical"

df["alert"] = df["predicted_probability"].apply(alert_level)
```

```python
def ai_summary(row):

    if row["alert"] == "Stable":
        return "Patient stable. Continue routine monitoring."

    if row["alert"] == "Watch":
        return "Moderate deterioration risk detected. Increase monitoring."

    if row["alert"] == "Warning":
        return "Clinical deterioration likely. Physician review recommended."

    return "Critical deterioration risk. Immediate intervention required."

df["ai_summary"] = df.apply(ai_summary, axis=1)
```

```python
df.to_csv(
    "E:/icu_project/models/ui_backend_data.csv",
    index=False
)

print("UI backend dataset created")
```
