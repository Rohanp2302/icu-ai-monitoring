```python
import pandas as pd
import numpy as np
import wfdb
import ast

path = r"E:/icu_project/data/ptbxl"

meta = pd.read_csv(path + "/ptbxl_database.csv")
meta["scp_codes"] = meta["scp_codes"].apply(ast.literal_eval)

def abnormal(code):
    if "NORM" in code and len(code) == 1:
        return 0
    return 1

meta["label"] = meta["scp_codes"].apply(abnormal)

signals = []
labels = []

for i,row in meta.iloc[:2000].iterrows():  # small subset

    sig,_ = wfdb.rdsamp(path + "/" + row["filename_lr"])

    signals.append(sig)
    labels.append(row["label"])

signals = np.array(signals)
labels = np.array(labels)

np.save("E:/icu_project/data/ecg_signals.npy", signals)
np.save("E:/icu_project/data/ecg_labels.npy", labels)

print("Saved ECG dataset:", signals.shape)
```
